Q4. What is ‘DataFrame’ in pandas and how is it different from pandas.series? Explain with an example.
Answer:

In Pandas, a DataFrame is a two-dimensional labeled data structure that consists of rows and columns. It is similar to a spreadsheet or a SQL table. Each column in a DataFrame is a Pandas Series object.

A Pandas Series is a one-dimensional labeled array that can hold any data type, such as integers, floats, strings, and so on. It is similar to a column in a spreadsheet or a database table.

Here's an example to illustrate the difference between a DataFrame and a Series:

dataset={'name':['swapnil','pratik','sudhanshu','krish'],
        'age':['23','45','44','55'],
         'game':['cricket','football','no','kabbadi']
        
        }

data=pd.DataFrame(dataset)

data1=dataset['name']

print(data)
print(data1)

Q5. What are some common functions you can use to manipulate data in a Pandas DataFrame? Can
you give an example of when you might use one of these functions?

Answer:

head() and tail(): These functions are used to view the top or bottom rows of a DataFrame. For example, df.head() will display the first five rows of the DataFrame df, while df.tail() will display the last five rows.

info(): This function is used to get information about the DataFrame, such as the data types of each column and the number of non-null values in each column.

describe(): This function is used to get summary statistics of each numerical column in the DataFrame, such as the mean, standard deviation, and quartiles.

loc[] and iloc[]: These functions are used to select specific rows and columns from the DataFrame. loc[] is used to select rows and columns by label, while iloc[] is used to select rows and columns by integer position.

groupby(): This function is used to group the rows of a DataFrame based on one or more columns, and then apply a function to each group. For example, df.groupby('category').mean() will group the rows of the DataFrame df by the "category" column and calculate the mean of each numerical column for each group.

drop(): This function is used to remove rows or columns from the DataFrame. For example, df.drop('column_name', axis=1) will remove the column with the name "column_name" from the DataFrame df.


Q6. Which of the following is mutable in nature Series, DataFrame, Panel?

In Pandas, both Series and DataFrame are mutable in nature, which means that we can modify their contents after they have been created.

A Series is a one-dimensional labeled array that can hold any data type, and a DataFrame is a two-dimensional labeled data structure with columns of potentially different types. Both of these data structures can be modified by changing their values, adding or deleting columns or rows, or performing other operations on them.
