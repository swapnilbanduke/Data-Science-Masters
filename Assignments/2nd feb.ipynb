**Q1 What are the characteristics of the tuples? Is tuple immutable?**

A tuple is an ordered, immutable and heterogeneous collection of elements in Python. Here are some key characteristics of tuples:

1) Order: Tuples maintain the order of elements and allow duplicate values.

2) Immutability: Tuples are immutable, which means once created, their elements cannot be changed or modified.

3) Heterogeneity: Tuples can contain elements of different data types.

4) Indexing: Elements of a tuple can be accessed using indexing, just like lists.

5) Access: Tuples can be accessed, but their elements cannot be modified or removed.

In summary, tuples are immutable collections that are ordered, and can contain elements of different data types.

**Q4. Explain the difference between the union() and update() methods for a set. Give an example of
each method.**

Answer:

The union and update methods are both used to add elements to a set, but they work in slightly different ways:

1) union: The union method returns a new set that is the union of the original set and another set, list, or any other iterable. The original set is not modified. If the elements being added already exist in the original set, they are not added again, as sets do not allow duplicate elements.
set1 = {1, 2, 3}
set2 = {3, 4, 5}

#Find the union of set1 and set2
union_set = set1.union(set2)

print(union_set)
#Output: {1, 2, 3, 4, 5}

2)update: The update method adds all the elements from another set, list, or any other iterable to the original set. The original set is modified in place and no new set is returned. If the elements being added already exist in the set, they are not added again, as sets do not allow duplicate elements.

set1 = {1, 2, 3}
set2 = {3, 4, 5}

#Add elements from set2 to set1
set1.update(set2)

print(set1)
#Output: {1, 2, 3, 4, 5}


In summary, the union method returns a new set that is the union of the original set and another set, while the update method adds elements to the original set in place.

**Q5. What is a dictionary? Give an example. Also, state whether a dictionary is ordered or unordered.**

Answer:

A dictionary in Python is an unordered collection of key-value pairs, where each key is unique. Dictionaries are commonly used to store and retrieve data efficiently, as they allow you to access values based on their associated keys.

Here's an example of a dictionary:

person = {'name': 'John', 'age': 32, 'address': '123 Main St'}

In this example, the keys 'name', 'age', and 'address' are associated with the values 'John', 32, and '123 Main St', respectively. You can access the values using square brackets ([]) and the keys:

It is important to note that dictionaries in Python are unordered. This means that the keys and values in a dictionary are not stored in any particular order, and the order may change as you add or remove elements from the dictionary. If we need to maintain the order of elements in a collection, we can use a list or a tuple instead.

**Q8. What are the three view objects in dictionaries? Use the three in-built methods in python to display

Answer:

In Python, dictionaries have three view objects: 
keys(), values(), and items(). 
These view objects are essentially a dynamic view of the keys, values, or key-value pairs of a dictionary.

1) keys()
dict1.keys()

output: dict_keys(['Sport', 'Teams'])

2) values()

dict1.values()
dict_values(['Cricket', ['India', 'Australia', 'England', 'South Africa', 'Sri Lanka', 'New Zealand']])

3) dict1.items()

dict_items([('Sport', 'Cricket'), ('Teams', ['India', 'Australia', 'England', 'South Africa', 'Sri Lanka', 'New Zealand'])])
